```python
pinkwink_web=pd.read_csv('C:/Users/joohee/Documents/DataScience/data/08. PinkWink Web Traffic.csv',
                         encoding='utf-8', thousands=',', names=['date', 'hit'], index_col=0)
pinkwink_web=pinkwink_web[pinkwink_web['hit'].notnull()]
pinkwink_web.head()
```

```python
pinkwink_web['hit'].plot(figsize=(12,4), grid=True);
```

```python
time=np.arange(0, len(pinkwink_web))  # 
traffic=pinkwink_web['hit'].values

fx=np.linspace(0, time[-1], 1000)  #=  () 
```

```python
def error(f, x, y):  #  error    
    return np.sqrt(np.mean((f(x)-y)**2))  # error     ^2 
```

```python
fp1 = np.polyfit(time, traffic, 1)  #1  # polyfit =    y,.. 
f1 = np.poly1d(fp1) # poly1d =    () 

f2p = np.polyfit(time, traffic, 2)  #2 
f2 = np.poly1d(f2p)

f3p = np.polyfit(time, traffic, 3)  #3 
f3 = np.poly1d(f3p)

f15p = np.polyfit(time, traffic, 15)  #15 
f15 = np.poly1d(f15p)

print(error(f1, time, traffic))
print(error(f2, time, traffic))
print(error(f3, time, traffic))
print(error(f15, time, traffic))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(time, traffic, s=10)

plt.plot(fx, f1(fx), lw=4, label='f1')
plt.plot(fx, f2(fx), lw=4, label='f2')
plt.plot(fx, f3(fx), lw=4, label='f3')
plt.plot(fx, f15(fx), lw=4, label='f15')

plt.grid(True, linestyle='-', color='0.75')

plt.legend(loc=2)
plt.show()
```

```python
df=pd.DataFrame({'ds':pinkwink_web.index, 'y':pinkwink_web['hit']})
df.reset_index(inplace=True)
df['ds']=pd.to_datetime(df['ds'], format="%y. %m. %d.")   #to_datetime =  
del df['date']

m=Prophet(yearly_seasonality=True)  #yearly_seaonality =  (prophet  )
m.fit(df);
```

```python
future=m.make_future_dataframe(periods=60)  #make_future_dataframe = 60  
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
