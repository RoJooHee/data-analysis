```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/joohee/Documents/DataScience/data/08. example_wp_R.csv')
df['y'] = np.log(df['y'])
```

```python
df['cap'] = 8.5
```

```python
m = Prophet(growth='logistic', daily_seasonality=True) # logistic=로그함수
m.fit(df)
```

```python
future=m.make_future_dataframe(periods=1826)
future['cap']=8.5
fcst=m.predict(future)
m.plot(fcst);
```

```python
forecast=m.predict(future)
m.plot_components(forecast);
```
