```python
pinkwink_web=pd.read_csv('C:/Users/joohee/Documents/DataScience/data/08. PinkWink Web Traffic.csv',
                         encoding='utf-8', thousands=',', names=['date', 'hit'], index_col=0)
pinkwink_web=pinkwink_web[pinkwink_web['hit'].notnull()]
pinkwink_web.head()
```

```python
pinkwink_web['hit'].plot(figsize=(12,4), grid=True);
```

```python
time=np.arange(0, len(pinkwink_web))  #시간축 만듦
traffic=pinkwink_web['hit'].values

fx=np.linspace(0, time[-1], 1000)  #회귀=데이터를 간단한 모델(함수)로 표현
```

```python
def error(f, x, y):  #참 값과 error값 비교해 모델의 적합성 확인
    return np.sqrt(np.mean((f(x)-y)**2))  # error 구할 때 음수 안나오게 제곱근^2로 계산함
```

```python
fp1 = np.polyfit(time, traffic, 1)  #1차 함수 # polyfit = 함수 만드는데 필요한 y절편,차수..등 계산
f1 = np.poly1d(fp1) # poly1d = 위에서 찾은 값들로 함수(다항식) 만듦

f2p = np.polyfit(time, traffic, 2)  #2차 함수
f2 = np.poly1d(f2p)

f3p = np.polyfit(time, traffic, 3)  #3차 함수
f3 = np.poly1d(f3p)

f15p = np.polyfit(time, traffic, 15)  #15차 함수
f15 = np.poly1d(f15p)

print(error(f1, time, traffic))
print(error(f2, time, traffic))
print(error(f3, time, traffic))
print(error(f15, time, traffic))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(time, traffic, s=10)

plt.plot(fx, f1(fx), lw=4, label='f1')
plt.plot(fx, f2(fx), lw=4, label='f2')
plt.plot(fx, f3(fx), lw=4, label='f3')
plt.plot(fx, f15(fx), lw=4, label='f15')

plt.grid(True, linestyle='-', color='0.75')

plt.legend(loc=2)
plt.show()
```

```python
df=pd.DataFrame({'ds':pinkwink_web.index, 'y':pinkwink_web['hit']})
df.reset_index(inplace=True)
df['ds']=pd.to_datetime(df['ds'], format="%y. %m. %d.")   #to_datetime = 날짜라고 선언
del df['date']

m=Prophet(yearly_seasonality=True)  #yearly_seaonality = 주기성이 연단위(prophet함수 사용 시)
m.fit(df);
```

```python
future=m.make_future_dataframe(periods=60)  #make_future_dataframe = 60일간의 데이터 예측
future.tail()
```

```python
forecast=m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
